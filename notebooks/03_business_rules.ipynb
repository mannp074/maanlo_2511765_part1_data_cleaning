```python
import pandas as pd
import numpy as np

df = pd.read_excel("cleaned_orders.xlsx")


# Discount cleaning

df["discount"] = pd.to_numeric(
    df["discount"],
    errors="coerce"
)

df["discount"] = df["discount"].fillna(0)

# Invalid discount
df["invalid_discount_flag"] = (
    (df["discount"] < 0)
    | (df["discount"] > 0.5)
)


# Order / payment flags

df["cancelled_order_flag"] = (
    df["order_status"]
    .astype(str)
    .str.lower()
    .eq("cancelled")
)

df["failed_payment_flag"] = (
    df["payment_status"]
    .astype(str)
    .str.lower()
    .eq("failed")
)

df["refunded_order_flag"] = (
    df["payment_status"]
    .astype(str)
    .str.lower()
    .eq("refunded")
)


# Date conversion

df["order_date"] = pd.to_datetime(df["order_date"])
df["ship_date"] = pd.to_datetime(df["ship_date"])

# Invalid shipping
df["invalid_shipping_flag"] = (
    df["ship_date"] < df["order_date"]
)

# Save
df.to_excel(
    "cleaned_orders.xlsx",
    index=False
)

print("="*50)
print("Invalid discounts:", df["invalid_discount_flag"].sum())
print("Cancelled orders:", df["cancelled_order_flag"].sum())
print("Failed payments:", df["failed_payment_flag"].sum())
print("Refunded orders:", df["refunded_order_flag"].sum())
print("Invalid shipping records:", df["invalid_shipping_flag"].sum())
print("="*50)
```
