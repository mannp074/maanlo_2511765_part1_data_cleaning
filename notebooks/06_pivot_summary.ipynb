```python
import pandas as pd

df = pd.read_excel("cleaned_orders.xlsx")


# Valid sales dataset
# Exclude cancelled orders and failed payments

valid_df = df[
    (~df["cancelled_order_flag"])
    &
    (~df["failed_payment_flag"])
    &
    (~df["refunded_order_flag"])
]


# 1 Sales and Profit by Region

pivot_region = (
    valid_df
    .pivot_table(
        index="region",
        values=["calculated_sales","calculated_profit"],
        aggfunc="sum"
    )
    .sort_values(
        by="calculated_sales",
        ascending=False
    )
)


# 2 Sales and Profit by Category/Subcategory

pivot_category = (
    valid_df
    .pivot_table(
        index=["category","sub_category"],
        values=["calculated_sales","calculated_profit"],
        aggfunc="sum"
    )
    .sort_values(
        by="calculated_sales",
        ascending=False
    )
)


# 3 Order Count by Ship Mode

pivot_ship_mode = (
    valid_df
    .pivot_table(
        index="ship_mode",
        values="order_id",
        aggfunc="count"
    )
    .rename(columns={"order_id":"order_count"})
    .sort_values(
        by="order_count",
        ascending=False
    )
)


# 4 Profit Margin by Segment

pivot_segment = (
    valid_df
    .pivot_table(
        index="segment",
        values="profit_margin",
        aggfunc="mean"
    )
    .sort_values(
        by="profit_margin",
        ascending=False
    )
)


# 5 Refunded / Cancelled / Failed by Region

issue_df = df[
    df["refunded_order_flag"]
    |
    df["cancelled_order_flag"]
    |
    df["failed_payment_flag"]
]

pivot_issue_region = (
    issue_df
    .pivot_table(
        index="region",
        values=[
            "refunded_order_flag",
            "cancelled_order_flag",
            "failed_payment_flag"
        ],
        aggfunc="sum"
    )
)


# 6 Monthly Sales Trend

pivot_month = (
    valid_df
    .pivot_table(
        index=["order_year","order_month"],
        values="calculated_sales",
        aggfunc="sum"
    )
)


# Export

with pd.ExcelWriter("pivot_summary.xlsx") as writer:

    pivot_region.to_excel(
        writer,
        sheet_name="Sales_Profit_By_Region"
    )

    pivot_category.to_excel(
        writer,
        sheet_name="Category_Subcategory"
    )

    pivot_ship_mode.to_excel(
        writer,
        sheet_name="Order_Count_Ship_Mode"
    )

    pivot_segment.to_excel(
        writer,
        sheet_name="Profit_Margin_Segment"
    )

    pivot_issue_region.to_excel(
        writer,
        sheet_name="Issue_Orders_By_Region"
    )

    pivot_month.to_excel(
        writer,
        sheet_name="Monthly_Sales_Trend"
    )

print("pivot_summary.xlsx created successfully")
```
