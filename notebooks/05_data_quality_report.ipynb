```python
import pandas as pd

df = pd.read_excel("cleaned_orders.xlsx")


# Missing value summary

missing_summary = (
    df.isnull()
    .sum()
    .reset_index()
)

missing_summary.columns = [
    "Column",
    "Missing_Count"
]

business_rule_fixes = pd.DataFrame({
    "Issue": [
        "Missing Region Filled with Unknown",
        "Missing Ship Mode Filled with Unknown"
    ],
    "Count": [
        (df["region"] == "Unknown").sum(),
        (df["ship_mode"] == "Unknown").sum()
    ]
})


# Duplicate summary

duplicate_summary = pd.DataFrame({
    "Metric":[
        "Exact Duplicate Rows Found",
        "Duplicate Order ID Rows"
    ],
    "Count":[
        20,
        63
    ]
})


# Invalid discount summary

invalid_discount_summary = (
    df[df["invalid_discount_flag"]]
)


# Date issue summary

date_issue_summary = (
    df[df["invalid_shipping_flag"]]
)


# Order status issue summary

order_status_summary = pd.DataFrame({
    "Metric":[
        "Cancelled Orders",
        "Failed Payments",
        "Refunded Orders"
    ],
    "Count":[
        df["cancelled_order_flag"].sum(),
        df["failed_payment_flag"].sum(),
        df["refunded_order_flag"].sum()
    ]
})


# Sales / profit mismatch summary

sales_profit_mismatch = df[
    (
        df["sales_difference"].abs() > 1
    )
    |
    (
        df["profit_difference"].abs() > 1
    )
]


# Final quality summary

final_summary = pd.DataFrame({
    "Quality_Flag":[
        "Clean",
        "Warning",
        "Invalid"
    ],
    "Count":[
        (df["data_quality_flag"]=="Clean").sum(),
        (df["data_quality_flag"]=="Warning").sum(),
        (df["data_quality_flag"]=="Invalid").sum()
    ]
})


# Export

with pd.ExcelWriter(
    "data_quality_report.xlsx"
) as writer:
    
    business_rule_fixes.to_excel(
    writer,
    sheet_name="Business_Rule_Fixes",
    index=False
    )
    
    missing_summary.to_excel(
        writer,
        sheet_name="Missing_Value_Summary",
        index=False
    )

    duplicate_summary.to_excel(
        writer,
        sheet_name="Duplicate_Summary",
        index=False
    )

    invalid_discount_summary.to_excel(
        writer,
        sheet_name="Invalid_Discount_Summary",
        index=False
    )

    date_issue_summary.to_excel(
        writer,
        sheet_name="Date_Issue_Summary",
        index=False
    )

    order_status_summary.to_excel(
        writer,
        sheet_name="Order_Status_Summary",
        index=False
    )

    sales_profit_mismatch.to_excel(
        writer,
        sheet_name="Sales_Profit_Mismatch",
        index=False
    )

    final_summary.to_excel(
        writer,
        sheet_name="Final_Quality_Summary",
        index=False
    )

print("data_quality_report.xlsx created successfully")
```
