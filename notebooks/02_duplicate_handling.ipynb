```python
import pandas as pd

# Load cleaned dataset
df = pd.read_excel("cleaned_orders.xlsx")

print("="*60)
print("ORIGINAL SHAPE")
print(df.shape)


# Exact Duplicate Rows

exact_duplicates = df[df.duplicated()]

print("\nExact duplicate rows found:", len(exact_duplicates))

# Save exact duplicates for inspection
if len(exact_duplicates) > 0:
    exact_duplicates.to_excel(
        "exact_duplicate_rows.xlsx",
        index=False
    )


# Duplicate order IDs

duplicate_order_rows = df[df.duplicated(
    subset=["order_id"],
    keep=False
)]

duplicate_order_count = duplicate_order_rows["order_id"].nunique()

print("\nRows having duplicate order_id:", len(duplicate_order_rows))
print("Number of unique duplicate order_ids:", duplicate_order_count)

# Save duplicate order IDs for review
if len(duplicate_order_rows) > 0:
    duplicate_order_rows.sort_values("order_id").to_excel(
        "duplicate_order_ids.xlsx",
        index=False
    )


# Remove Exact Duplicates

df_no_duplicates = df.drop_duplicates()

records_removed = len(df) - len(df_no_duplicates)

print("\nRecords removed:", records_removed)

# Save updated cleaned file
df_no_duplicates.to_excel(
    "cleaned_orders.xlsx",
    index=False
)

print("\nUpdated Shape:", df_no_duplicates.shape)

print("\nTask 4 - Duplicate Handling Completed")
print("="*60)
```
